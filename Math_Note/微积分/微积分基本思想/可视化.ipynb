```python
# Visualization for Green's theorem and Gauss's theorem
import numpy as np
import matplotlib.pyplot as plt

# -------- Green's Theorem Visualization --------
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

# Vector field F = (-y, x) -> rotational field
U = -Y
V = X

theta = np.linspace(0, 2 * np.pi, 400)
cx = np.cos(theta)
cy = np.sin(theta)

plt.figure()
plt.quiver(X, Y, U, V)
plt.plot(cx, cy)
plt.gca().set_aspect("equal")
plt.title("Green's Theorem: Curl inside equals circulation on boundary")
plt.xlabel("x")
plt.ylabel("y")
plt.show()


# -------- Gauss's Theorem Visualization --------
x = np.linspace(-2, 2, 8)
y = np.linspace(-2, 2, 8)
z = np.linspace(-2, 2, 8)
X, Y, Z = np.meshgrid(x, y, z)

# Vector field F = (x, y, z) -> divergent field
U = X
V = Y
W = Z

fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.quiver(X, Y, Z, U, V, W, length=0.3, normalize=True)

# Sphere boundary
phi = np.linspace(0, np.pi, 30)
theta = np.linspace(0, 2 * np.pi, 30)
phi, theta = np.meshgrid(phi, theta)
xs = np.sin(phi) * np.cos(theta)
ys = np.sin(phi) * np.sin(theta)
zs = np.cos(phi)
ax.plot_wireframe(xs, ys, zs)

ax.set_title("Gauss's Theorem: Divergence inside equals flux through boundary")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
plt.show()
```

# Gree formula

```python
import numpy as np
import matplotlib.pyplot as plt

# 
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)

#  F = (-y, x)
U = -Y
V = X

# 
plt.figure(figsize=(8, 8))
plt.quiver(X, Y, U, V, color="blue", scale=20)
plt.title(" F = (-y, x)  quiver \n curl F = 2 ()")

#  C
theta = np.linspace(0, 2 * np.pi, 100)
plt.plot(np.cos(theta), np.sin(theta), "r-", linewidth=3, label=" C ()")

#  D
plt.fill(np.cos(theta), np.sin(theta), alpha=0.2, color="green", label=" D")

plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.legend()
plt.grid(True)
plt.show()

#  2 2 *  * 1^2 = 2
print(" = 2 ", 2 * np.pi)
print(" =   2 = 2 ", 2 * np.pi)
```

# Gauss Formula

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")

# 
u = np.linspace(0, 2 * np.pi, 30)
v = np.linspace(0, np.pi, 20)
x = np.outer(np.cos(u), np.sin(v))
y = np.outer(np.sin(u), np.sin(v))
z = np.outer(np.ones(np.size(u)), np.cos(v))

# 
ax.plot_surface(x, y, z, color="lightblue", alpha=0.3, linewidth=0)

#  F = (x, y, z) |F| = 1 ()
# 
scale = 1.5
for i in range(0, len(u), 5):
    for j in range(0, len(v), 4):
        px = np.cos(u[i]) * np.sin(v[j])
        py = np.sin(u[i]) * np.sin(v[j])
        pz = np.cos(v[j])
        ax.quiver(
            px, py, pz, px, py, pz, length=0.5, color="red", arrow_length_ratio=0.3
        )

ax.set_title(
    " F = (x, y, z) \n div F = 3 = 3  (4/3 ) = 4\n = 4 ()"
)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_box_aspect([1, 1, 1])
plt.show()

# 
print(" = 4 ", 4 * np.pi)
print(" = 3   = 4 ", 4 * np.pi)
```

```python
#  Green   Gauss 
# 
#  matplotlib 

import numpy as np
import matplotlib.pyplot as plt


# ========== Green  ==========
#  F = (P, Q)
def P(x, y):
    return -y


def Q(x, y):
    return x


# 
x = np.linspace(-1.2, 1.2, 25)
y = np.linspace(-1.2, 1.2, 25)
X, Y = np.meshgrid(x, y)
U = P(X, Y)
V = Q(X, Y)

# 
theta = np.linspace(0, 2 * np.pi, 400)
cx = np.cos(theta)
cy = np.sin(theta)

plt.figure()
plt.quiver(X, Y, U, V, alpha=0.8)
plt.plot(cx, cy, linewidth=2)
plt.gca().set_aspect("equal")
plt.title("Green Formula: Circulation  Curl Integral")
plt.xlabel("x")
plt.ylabel("y")
plt.show()


# ========== Gauss  ==========
#  F = (x, y)
def Fx(x, y):
    return x


def Fy(x, y):
    return y


U2 = Fx(X, Y)
V2 = Fy(X, Y)

plt.figure()
plt.quiver(X, Y, U2, V2, alpha=0.8)
plt.plot(cx, cy, linewidth=2)
plt.gca().set_aspect("equal")
plt.title("Gauss Formula (Divergence Theorem): Flux  Divergence Integral")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
#  Gauss 
#  3D  + + 

import numpy as np
import matplotlib.pyplot as plt


# ========= 1.  F = (x, y, z) =========
def Fx(x, y, z):
    return x


def Fy(x, y, z):
    return y


def Fz(x, y, z):
    return z


# ========= 2.  =========
grid = np.linspace(-1.2, 1.2, 6)
X, Y, Z = np.meshgrid(grid, grid, grid)

U = Fx(X, Y, Z)
V = Fy(X, Y, Z)
W = Fz(X, Y, Z)

# ========= 3.  V =========
phi = np.linspace(0, np.pi, 40)
theta = np.linspace(0, 2 * np.pi, 80)
phi, theta = np.meshgrid(phi, theta)

R = 1.0
Xs = R * np.sin(phi) * np.cos(theta)
Ys = R * np.sin(phi) * np.sin(theta)
Zs = R * np.cos(phi)

# ========= 4.  =========
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# 
ax.quiver(X, Y, Z, U, V, W, length=0.15, normalize=True, alpha=0.6)

# 
ax.plot_surface(Xs, Ys, Zs, rstride=2, cstride=2, linewidth=0, alpha=0.35)

# 
ax.set_box_aspect([1, 1, 1])
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("Gauss Theorem (3D): Flux through Surface  Volume Divergence")

plt.show()
```
